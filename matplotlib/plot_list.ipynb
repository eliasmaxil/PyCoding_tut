```python
import matplotlib.pyplot as plt
from matplotlib import gridspec
import numpy as np
```

```python
def nsubplots(sp=6, cols = 2, **kwargs):
    rows = int(np.ceil(sp/cols))
    gs = gridspec.GridSpec(rows, cols)
    c = 0
    fig = plt.figure(**kwargs)

    for i in range(sp):
        x = np.arange(0,5) 
        y = np.random.rand(5) * (1+c*.1)
        ax = fig.add_subplot(gs[i])
        ax.plot(x,y)
        ax.set_xlabel('x')
        ax.set_ylabel('y')
        ax.set_title(str(c), x=0.5, y=.8)
        ax.grid(axis='y', alpha=0.55)
        c += 1
    return fig, ax

f, a = nsubplots(figsize=(16,6))
# f.savefig('temp.png')
# f
```
