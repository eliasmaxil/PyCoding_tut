```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
# A very delayed ts
t_pe = [dt.datetime(2023,1,24,15,0,0), dt.datetime(2023,12,24,17,0,50)]
idx_pe = pd.date_range(start=t_pe[0], end=t_pe[1], freq='60T', tz='utc')
cols_pe = ['pe1','pe2','pe3']
df_pe = pd.DataFrame(np.random.rand(len(idx_pe),3)*100, \
    index=idx_pe, columns=cols_pe)
df_pe
```

```python
# A not really live ts
tnow = dt.datetime.utcnow()
tnow = dt.datetime(tnow.year,tnow.month,tnow.day,tnow.hour,tnow.minute+12)
t_sph = [tnow-dt.timedelta(days=9, hours=3), tnow-dt.timedelta(days=2, hours=13)]
idx_sph = pd.date_range(start=t_sph[0], end=t_sph[1], freq='60T', tz='utc')
df_sph = pd.DataFrame(np.zeros(len(idx_sph)),index=idx_sph, columns=['dummy'])
df_sph
```

```python
tnow = dt.datetime.utcnow()
tnow = dt.datetime(tnow.year,tnow.month,tnow.day,tnow.hour,tnow.minute)
t_live = [tnow - dt.timedelta(days=3), tnow]
idx_live = pd.date_range(start=t_live[0], end=t_live[1], freq='60T', tz='utc')
df_live = pd.DataFrame(np.zeros(len(idx_live)),index=idx_live, columns=['dummy'])
df_live
```
